## 通用工具

### 学习器类型

```python
import sklearn.preprocessing as preprocessing

log_transformer = preprocessing.FunctionTransformer(
    func = np.log1p
)
```

```python
x = [
    [1, 3, 4, 9, 10],
    [2, 1, 0, 0, 0]
]
x_log = log_transformer.fit_transform(x)
x_log
```

```python
import pandas as pd

df = pd.DataFrame({
    "age": [25, 40, 35, 50],
    "income": [70000, 120000, 90000, 150000],
    "buy": [0, 1, 1, 1]
})

def add_income_per_age(X):
    X = X.copy()
    X["income_per_age"] = X["income"] / X["age"]
    return X

feature_adder = preprocessing.FunctionTransformer(
    func = add_income_per_age,
    validate = False
)

feature_adder.fit_transform(df)
```
